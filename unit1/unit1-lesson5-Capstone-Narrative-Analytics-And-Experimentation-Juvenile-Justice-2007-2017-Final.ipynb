## Thinkful Research Proposal:

### Problem:
Despite a dramatic decrease in juvenile probation contacts between 2007 - 2017, there continues to appear to be significant differences across race in the way youth enter the juvenile criminal justice system in California.

Since detention removes youth from their families, their schools, and their communities, it disrupts the integrity of these natural supports in a youth's life.  These supports are critical environmental factors crucial to developing and engaging in pro-social behavior.

A youth's removal from their natural supports could have adverse and unintended consequences that actually contribute to criminal thinking and behavior.

From a data perspective, it is not clear what kinds of concrete structural changes could be made to the juvenile justice system that might address the differences in the way the institution applies the law.

Equally important, it is not yet clear what kind of adverse and unintended consequences may occur due to imprisonment of youth and the resulting separations from family, school, and community. The juvenile justice data that is available does not allow for this kind of analysis.

The data suggest that more than 1 in 5 Black youth in California are detained in a Secure Facility before having their cases dismissed.  These Secure Facility Detentions that ultimately result in a dismissed "Disposition" may be an area to target in beginning to address the problem.

### Hypothesis:
When a youth is detained in a Secure Facility at first contact with law enforcement, there is an increase in likelihood of continued contacts with the juvenile justice system.

### Potential Solution:
Randomly assign youth "13 and Under" at first law enforcement contact who would typically be detained in a Secure Facility (for a Court Referral Type) to a non-secure facility based on clear, selected criteria (for example, assessed to able to function at a non-secure facility without being a danger to self or to others).

### Implementation
1. Determine sample size large enough that allows ability to calculate significant differences over time.
- Identify 3-5 Counties across State that are varied in size and resources.
- Develop criteria for entry into intervention in consultation with Counties (probation and non-secure facilities) to ensure random assignments are safe for staff and youth, they are consistently applied across sites, and work within logistical resources of intervention sites.
- Develop clear protocol and policy to use for non-secure facility youth in case their presentation indicates need to return to secure facility housing
- Develop/modify existing tracking systems that can be scalable to follow unique youth in intervention across time
- Roll out intervention to 3-5 County sites for 2 months to ensure fidelity to policies/procedures identified and safety of youth and staff remains at expected levels for intervention
- Evaluate safety and potential problems to determine if continuing intervention is appropriate. If so, continue to roll out intervention to include five new Counties per month ~ 11 month Roll out process (58 Counties total).
- Begin reviewing data at quarterly basis until determined sample size is reached.
- If rollout meets success metrics after 12 months following assignments to intervention at determined sample size, follow all assigned to non-secure vs assigned to secure facilitities youth to evaluate metrics

### Metrics
- Primary Metric: Frequency of law enforcement contacts following first contact and random assignment
- Secondary Metric: Sustained offense levels (Felony vs Misdemeanor offenses)
- Tertiary Metric: Duration of time in custody

Intervention success will mean significant difference between intervention Non-Secure group vs. "as usual" Secure Facility group in specified metrics.

### Policy Recommendations:
In a real-world context, I would actually propose tracking youth first based on unique IDs.  There are multiple assumptions made with the proposed solution.  There are potentially significant and large lurking variables that would likely set up this potential solution for failure.

For example, while detention in a Secure Facility might be potentially traumatic, it could also be safer than the environment out of which the youth typically resides.  For a child or youth living in a dangerous or chaotic environment, a Secure Facility might represent stability, consistency, and containment - and be a positive factor for that child.

In this situation, a non-secure facility might actually be experienced as less safe than a Secure Facility.  Understanding the conditions of the children and youth who are being detained may offer more insight into what kind of intervention may be better suited to achieve the end result.

Ultimately, creating or developing a data infrastructure to track youth through the juvenile justice system that integrated with other State and County systems (probation, education, legal, social welfare, etc) would make it possible to glean more meaningful insights.

This infrastructure would be an important and necessary step in beginning to tease out the complexity to get data driven insights that could support policy that makes communities safer and ensures all youth are treated fairly and justly.

## Basic contextual overview to understand the data:

### Stage 1: Referral action

When a youth has contact with law enforcement (they allegedly commit a crime), the contact is considered a "Referral" action.  There are several potential inital outcomes of this "Referral".  The youth could be:

- Not detained
- Detained under home supervision
- Detained in a non-secure facility
- Detained in a secure facility

If a youth is not initially detained, they may be given a ticket or citation to report to a probation department where they will discuss their case with a probation officer. If they are detained, a decision must be made within 72 hours regarding their continued detention.

Many youth are released to parents or guardians within this 72 hour period.  Youth may also be sent to a crisis assessment center, a shelter or group home.  Their release is "conditional" at this time.

If a youth is initially detained, then there are various "dispositions" or potential outcomes for the Referral.  The Referral can be classified as:

- Closed
- Deported
- Direct File - Adult Court
- Dismissed
- Diversion
- Informal Probation
- Traffic Court
- Transferred

### Stage 2: Court action

If it is decided that detention needs to be continued beyond the 72 hours, then the "Referral" becomes a "Court" action petition.  The District Attorney files legal charges and the case is presented to the court with the youth present.

Once the Referral becomes a Court petition, there are several potential outcomes:

- Deferred Entry of Judgement
- Deported
- Dismissed
- Diversion
- Informal Probation (654.2 WI)
- Non-Ward Probation (725a WI)
- Remanded to Adult Court
- Transferred
- Wardship
- Wardship (Other)

```python
#Visualize new and subsequent petitions for boys
# Setting the positions and width for the bars
pos = list(range(len(new_13_male_counts['Gender']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(20,10))

# Create a bar with BOYS data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        new_13_male_counts['Gender'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=new_13_male_counts['Gender']) 

# Create a bar with GIRLS data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        subsequent_13_male_counts['Gender'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=subsequent_13_male_counts['Gender']) 

# Set the y axis label
ax.set_ylabel('Number of Referrals')

# Set the chart's title
ax.set_title('2007 - 2017: Boys Age 13 and Under Referred for a "New" vs "Subsequent" Petition')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['Report_year'].unique())

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(new_13_male_counts['Gender'] + subsequent_13_male_counts['Gender'])] )

# Adding the legend and showing the plot
plt.legend(['New', 'Subsequent'], loc='upper left')
plt.grid()
plt.show()
```
