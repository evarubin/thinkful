```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Distribution 1: BINOMIAL distribution = binomial_dist

```python
#Distribution 1: BINOMIAL distribution = np.random.binomial(1, .5, 1000)

binomial_dist = np.random.binomial(1, .5, 100)
```

```python
#Distribution 1:2 plot Binomial Histogram

plt.hist(binomial_dist);
```

```python
# Do a qq plot

# Create the normal distribution to compare against
norm = np.random.normal(0, 1, 100)

# Sort the values in ascending order.
norm.sort()

# Sort the variable values
binomial_dist.sort()

# Plot the variable against norm in qqplots.
plt.plot(norm, binomial_dist, "o")
plt.show() 
```

```python
#Distribution 1:3 mean and standard deviation
binomial_dist_mean = binomial_dist.mean()
binomial_dist_mean
```

```python
binomial_dist_std = binomial_dist.std()
binomial_dist_std
```

```python
#Distribution 1:3 mean and standard deviation vertical lines on the histogram

plt.hist(binomial_dist);
plt.axvline(binomial_dist_mean, color='c', linestyle='dotted', linewidth=2)
plt.axvline(binomial_dist_std, color='b', linestyle='solid', linewidth=2)

plt.show()
```

```python
#Distribution 1:4 Evaluate whether the descriptive statistics provided useful information about the variable.
#Can you identify any common characteristics of the distributions that could be usefully described using the mean
#and/or standard deviation, versus the ones that could not?

#Non-normal distribution, no relationship between the variables.
```

## Distribution 3: POISSON distribution = poisson_dist

```python
#Distribution 3: POISSON distribution

# What is the lambda of Poisson?
#Counts of events, based on the Poisson distribution, is a frequently encountered model in medical research
#Examples of this are number of falls, asthma attacks, number of cells, and so on.
#The Poisson parameter Lambda (λ) =
#the total number of events (k) divided by the number of units (n) in the data (λ = k/n).

poisson_dist = np.random.poisson(3, 100)
```

```python
#Distribution 3:2 plot Poisson Histogram
plt.hist(poisson_dist);
```

```python
# Do a qq plot

# Create the normal distribution to compare against
norm = np.random.normal(0, 1, 100)

# Sort the values in ascending order.
norm.sort()

# Sort the variable values
poisson_dist.sort()

# Plot the variable against norm in qqplots.
plt.plot(norm, poisson_dist, "o")
plt.show() 
```

```python
#Distribution 3:3 mean and standard deviation
poisson_dist_mean = poisson_dist.mean()
poisson_dist_mean
```

```python
poisson_dist_std = poisson_dist.std()
poisson_dist_std
```

```python
#Distribution 3:3 mean and standard deviation vertical lines on the histogram
plt.hist(poisson_dist);
plt.axvline(poisson_dist_mean, color='c', linestyle='dotted', linewidth=2)
plt.axvline(poisson_dist_std, color='b', linestyle='solid', linewidth=2)

plt.show()
```

```python
#Distribution 3:4 Evaluate whether the descriptive statistics provided useful information about the variable.
#Can you identify any common characteristics of the distributions that could be usefully described using the mean
#and/or standard deviation, versus the ones that could not?

#An approsimately normal distribution.
```
